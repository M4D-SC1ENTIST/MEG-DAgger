```python
import numpy as np
import pandas as pd
import seaborn as sns

import matplotlib
import matplotlib.pyplot as plt
import cv2
```

# General1

## at1

```python
general1_at1_single_without_filter = pd.read_csv('training_performance_comp/new/General1/at1/singe_without_filter.csv')
general1_at1_single_with_filter = pd.read_csv('training_performance_comp/new/General1/at1/single_with_filter.csv')
general1_at1_multi_with_filter = pd.read_csv('training_performance_comp/new/General1/at1/multi_with_filter.csv')

general1_at1_multi_with_filter['Method'] = 'MEG-DAgger'
general1_at1_single_with_filter['Method'] = 'HG-DAgger with filter'
general1_at1_single_without_filter['Method'] = 'HG-DAgger'
general1_at1_concat = pd.concat([general1_at1_multi_with_filter, general1_at1_single_with_filter, general1_at1_single_without_filter], ignore_index=True)


matplotlib.rcParams.update({'font.size': 11})

general1_at1_fig, general1_at1_axarr = plt.subplots(1, 2, figsize=(10, 3.5))
hue_order = ['MEG-DAgger', 'HG-DAgger with filter', 'HG-DAgger']

general1_at1_a = sns.lineplot(x='model_idx', y='avg_overtake_rate', hue="Method", data=general1_at1_concat, ax=general1_at1_axarr[0], hue_order=hue_order, palette=['#384871', "#5EA69C", "#CD4432"], legend=False)
general1_at_b = sns.lineplot(x='model_idx', y='avg_collision_rate', hue="Method", data=general1_at1_concat, ax=general1_at1_axarr[1], hue_order=hue_order, palette=['#384871', "#5EA69C", "#CD4432"], legend=False)

general1_at1_a.set_xlabel('Number of Rollouts')
general1_at1_a.set_ylabel('Overtake Rate')

general1_at_b.set_xlabel('Number of Rollouts')
general1_at_b.set_ylabel('Collision Rate')
```
