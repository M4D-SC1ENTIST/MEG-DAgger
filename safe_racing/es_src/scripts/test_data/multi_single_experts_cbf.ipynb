```python
import numpy as np
import pandas as pd
import seaborn as sns

import matplotlib
import matplotlib.pyplot as plt
import cv2
```

```python
multi_experts_cbf = pd.read_csv('multi_weak_expert_cbf.csv')
single_expert_5392_cbf = pd.read_csv('single_weak_expert5392_cbf.csv')
single_expert_no_cbf = pd.read_csv('single_weak_expert_no_cbf.csv')

multi_experts_cbf['Method'] = 'MEG-DAGGER'
single_expert_5392_cbf['Method'] = 'Single expert with data filter'
single_expert_no_cbf['Method'] = 'Single expert without data filter'

comparison_df = pd.concat([multi_experts_cbf, 
                           single_expert_5392_cbf,
                           single_expert_no_cbf], ignore_index=True)
```

```python
matplotlib.rcParams.update({'font.size': 10})

fig, axarr = plt.subplots(1, 2, figsize=(10, 4.2))
hue_order = ['MEG-DAGGER', 'Single expert with data filter', 'Single expert without data filter']
a = sns.lineplot(x='model_idx', y='avg_overtake_rate', hue="Method", data=comparison_df, ax=axarr[0], hue_order=hue_order, palette=['#384871', "#5EA69C", "#CD4432"], legend=False)
b = sns.lineplot(x='model_idx', y='avg_collision_rate', hue="Method", data=comparison_df, ax=axarr[1], hue_order=hue_order, palette=['#384871', "#5EA69C", "#CD4432"], legend=True)

a.set_xlabel('Iterations')
a.set_ylabel('Overtake Rate')

b.set_xlabel('Iterations')
b.set_ylabel('Collision Rate')

b.legend(loc='best', bbox_to_anchor=(0.2, 0.9))
```
